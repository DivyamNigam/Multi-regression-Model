```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\divya\Downloads\cleaned_data.xlsx")
```

```python
df.dropna(inplace=True)
x = df.drop(columns=['Listing ID', 'List Date', 'Sold Date', 'Price'])
y= df['Price']
c = df['Bedrooms']
lr = LinearRegression()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
lr.fit(x_train, y_train)
```

```python
np.isnan(y).any()
pred_modle= lr.predict(x)
```

```python
lr.score(x_test, y_test)
```

```python
# Initialize the plot
plt.figure(figsize=(10, 6))

# Define a color palette
colors = sns.color_palette("husl", len(x.columns))

# Loop through each column and plot it
for i, column in enumerate(x.columns):
    plt.scatter(x[column], y, color=colors[i], label=column, alpha=0.6)

# Add title and labels
plt.title('Scatter Plot of Multiple Columns')
plt.xlabel('Index')
plt.ylabel('Value')
# Show the plot
plt.show()
```

```python
# Initialize the plot
plt.figure(figsize=(10, 6))

# Define a color palette
colors = sns.color_palette("husl", len(x.columns))

# Loop through each column and plot it
for i, column in enumerate(x.columns):
    plt.scatter(x[column], y, color=colors[i], label=column, alpha=0.6)
    plt.plot(x[column], pred_modle, color='red', linewidth=2)

# Add title and labels
plt.title('Scatter Plot of Features with Regression Lines')
plt.xlabel('Feature Value')
plt.ylabel('Price')

# Add legend
plt.legend()

# Show the plot
plt.show()

```
